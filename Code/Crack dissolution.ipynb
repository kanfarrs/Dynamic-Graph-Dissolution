```python
from dgd import crack #will have to install: karstnet, pyvista, open3d, kmapper, mplstereonet, kneebow, multiprocess (and maybe others depending on env)
import numpy as np 
```

## Single Crack dissolution

```python
#crackDiss params
num_years = 20000               #[years]
dt = 10                         #[years]
init_alpha = 0.02               #[cm]
init_beta = 100                 #[cm]
length = 1e4                    #[cm] if larger, may have to load spatial array instead of initializing it
hgrad = 0.01                    #[unitless]

#run crackDiss
diss = crack(num_years, dt)
grid = diss.crackDiss(length, init_alpha, init_beta, hgrad)
```

```python
#plot results
viewStamps = np.array([0, 10000, 17000, 17250, 17500])   #[years]
diss.plotCrack(grid, viewStamps)
```
